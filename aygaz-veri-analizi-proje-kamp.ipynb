# 🛠️ **Rastgele Eksik Veri Ekleme Fonksiyonu**

Bu bölümde, bir veri çerçevesine rastgele eksik değerler eklemek için bir fonksiyon oluşturulmuş ve uygulanmıştır. Eksik veri ekleme, özellikle makine öğrenimi ve veri analitiği projelerinde, eksik verilerle başa çıkma yöntemlerini test etmek için kullanışlıdır.

----------

## **🔍 Fonksiyonun Amacı**

### **Fonksiyon: `add_random_missing_values_to_all_columns`**

Bu fonksiyonun görevi, belirtilen oran (%5) doğrultusunda bir veri çerçevesindeki tüm sütunlara rastgele eksik (NaN) değerler eklemektir.

### **Fonksiyonun Parametreleri:**

-   **`dataframe`**: Üzerinde işlem yapılacak Pandas veri çerçevesi.
-   **`missing_rate_per_col`**: Her sütunda ne kadar eksik veri eklenmesi gerektiğini belirten oran (varsayılan olarak %5).
-   **`seed`**: Rastgelelik için sabit bir başlangıç noktası belirlemek ve tekrar edilebilir sonuçlar almak için kullanılır.

----------

## **💡 Yapılan İşlemler**

1.  **Orijinal Veriyi Koruma:**
    
    -   Veri çerçevesinin bir kopyası oluşturularak orijinal veri korunur.
2.  **Rastgelelik Kontrolü:**
    
    -   `random.seed()` ve `np.random.seed()` kullanılarak rastgele işlemler sabitlenir. Bu, fonksiyonun tekrar çalıştırıldığında aynı sonuçları üretmesini sağlar.
3.  **Sütun Türlerini Dönüştürme:**
    
    -   Eksik değerler (NaN) yalnızca float türündeki verilere atanabileceği için, integer türündeki sütunlar float’a dönüştürülür.
4.  **Eksik Veri Ekleme:**
    
    -   Her sütun için:
        -   Eksik veri eklenmesi gereken satır sayısı hesaplanır.
        -   Rastgele seçilen satırların indekslerine NaN atanır.

----------

## **📊 Uygulama: Los Angeles Suç Verisi**

Fonksiyon, **Los Angeles suç verisi** üzerinde şu şekilde uygulanmıştır:

-   Her sütuna toplam verinin %5'i kadar eksik veri eklenmiştir.
-   Bu adım, eksik veri analizi ve doldurma yöntemlerini test etmek için kullanışlıdır.

```python
def add_random_missing_values_to_all_columns(dataframe: pd.DataFrame,
                                             missing_rate_per_col: float = 0.05,
                                             seed: int = 42) -> pd.DataFrame:
    """Adds random missing values to every column in a DataFrame.
    
    Args:
        dataframe (pd.DataFrame): DataFrame to be processed.
        missing_rate_per_col (float): Percentage of missing value rate in float format for each column. Defaults 0.05
        seed (int): Random seed for reproducibility.
        
    Returns:
        pd.DataFrame: DataFrame with added missing values.
    """
    # Copying the dataframe to prevent changing the original data
    df_missing = dataframe.copy()

    # Set seed for reproducibility
    random.seed(seed)
    np.random.seed(seed)

    # Convert integer columns to float to allow NaN values
    for col in df_missing.select_dtypes(include=['int']).columns:
        df_missing[col] = df_missing[col].astype(float)

    # Iterate over each column to add missing values
    for col in df_missing.columns:
        # Number of missing values to add in this column
        num_missing = int(len(df_missing) * missing_rate_per_col)
        
        # Generate random row indices to set as NaN
        missing_indices = np.random.choice(df_missing.index, num_missing, replace=False)
        
        # Set those indices to NaN in the respective column
        df_missing.loc[missing_indices, col] = np.nan

    return df_missing

# Veri setine her sütuna eksik veri eklemek için
crime_2020_to_present_data = add_random_missing_values_to_all_columns(
    dataframe=crime_2020_to_present_data,
    missing_rate_per_col=0.05  # Her sütuna %5 eksik veri ekleniyor
)
```

# **🔍 Veri Çerçevesi Kontrol Fonksiyonu: `check_df`**

Bu bölümde, bir veri çerçevesini detaylı şekilde analiz etmek ve düzenli bir biçimde görüntülemek için bir fonksiyon oluşturulmuştur. **`check_df`**, veri çerçevesi hakkında temel bilgileri, eksik değerleri ve istatistiksel özetleri kullanıcı dostu bir şekilde sunar. 📊✨

----------

## **🛠️ Fonksiyonun Amacı ve Yapılan İşlemler**

### 🎯 **Fonksiyonun Amacı**

`check_df` fonksiyonu, bir veri çerçevesinin:

-   Boyutlarını,
-   Sütun türlerini,
-   İlk ve son birkaç satırını,
-   Eksik değerlerini,
-   İstatistiksel özetini görüntülemeyi sağlar.

### 🔧 **Fonksiyonda Yapılan İşlemler:**

1.  **Genel Bilgiler (`info`)**
    
    -   Veri çerçevesi hakkında sütun bazında **non-null (boş olmayan)** değer sayısı ve veri türleri (`dtype`) gibi bilgiler tablolaştırılarak düzenli bir şekilde gösterildi.
2.  **Boyut Bilgisi (`shape`)**
    
    -   Veri çerçevesinin satır ve sütun sayısı kullanıcıya bildirildi.
3.  **Sütun Türleri (`dtypes`)**
    
    -   Tüm sütunların veri türleri (`int`, `float`, `object` gibi) görüntülendi.
4.  **İlk ve Son Satırlar (`head` ve `tail`)**
    
    -   Veri çerçevesinin ilk ve son birkaç satırı gösterilerek genel yapı hakkında bilgi sağlandı.
5.  **Eksik Değerler (`isnull`)**
    
    -   Sütun bazında eksik değerlerin toplamı hesaplanıp düzenli bir şekilde görüntülendi.
6.  **İstatistiksel Özet (`describe`)**
    
    -   Sayısal sütunlara ait istatistiksel özetler (ortalama, standart sapma, min, max gibi) tablolaştırıldı.

----------

## **📊 Fonksiyonun Uygulaması: Los Angeles Suç Verisi**

-   **Veri Yükleme:** `pd.read_csv()` ile **Los Angeles suç verisi** yüklendi.
-   **Fonksiyon Kullanımı:** `check_df` fonksiyonu kullanılarak veri çerçevesine dair:
    -   Yapısal bilgiler,
    -   Eksik değer analizi,
    -   Sayısal özetler kolayca görüntülendi.

```python
def check_df(dataframe, head=5):
    # Info - Düzenli tablo halinde gösterim
    buffer = StringIO()
    dataframe.info(buf=buffer)
    info_str = buffer.getvalue()

    # Info'yu tablolaştırma
    info_lines = info_str.splitlines()
    col_info = []
    columns = ['#', 'Column', 'Non-Null Count', 'Dtype']  # Sütun isimlerini dinamik hale getiriyoruz

    for line in info_lines[5:]:
        if line.strip():
            cols = line.split(maxsplit=3)  # Sadece 4 sütun ayırarak fazla bölünmeyi engelliyoruz
            if len(cols) == 4:  # Eğer tüm sütunlar varsa ekliyoruz
                col_info.append(cols)

    # DataFrame oluşturma ve görüntüleme
    info_df = pd.DataFrame(col_info, columns=columns)
    display(HTML("<h3>Info</h3>"))
    display(info_df)

    # Shape
    display(HTML(f"<h3>Shape</h3><p>{dataframe.shape}</p>"))
    
    # Types
    display(HTML("<h3>Types</h3>"))
    types_df = pd.DataFrame({'Column': dataframe.columns, 'Dtype': dataframe.dtypes}).transpose()
    display(types_df)
    
    # Head
    display(HTML("<h3>Head</h3>"))
    display(dataframe.head(head))
    
    # Tail
    display(HTML("<h3>Tail</h3>"))
    display(dataframe.tail(head))
    
    # NA (Null Values)
    display(HTML("<h3>NA (Null Values)</h3>"))
    na_df = dataframe.isnull().sum().to_frame().T
    na_df.columns.name = None  # Sütun isimlerinden `name` özelliğini temizledik
    display(na_df)
    
    # Describe
    display(HTML("<h3>Describe</h3>"))
    describe_df = dataframe.describe().transpose()
    describe_df.columns.name = None  # Sütun isimlerinden `name` özelliğini temizledik
    display(describe_df)

# Crime data çerçevesini kontrol etme
crime_2020_to_present_data = pd.read_csv("/kaggle/input/los-angeles-crime-data-from-2000/Crime_Data_from_2020_to_Present.csv")
check_df(crime_2020_to_present_data)
```

# 🛠️ **Sütun Türlerini Otomatik Tespit Etme: `grab_col_names`**

Bu fonksiyon, bir veri çerçevesindeki sütunları **kategorik**, **sayısal**, ve özel kategorilere ayırmak için geliştirilmiştir. Veri analizi ve modelleme süreçlerinde, sütun türlerini belirlemek önemli bir adımdır ve bu fonksiyon bu işlemi otomatikleştirir. 🔍✨

----------

## **🔍 Fonksiyonun Amacı**

**`grab_col_names`**, veri çerçevesindeki sütunları özelliklerine göre üç ana kategoriye ayırır:

1.  **Kategorik Sütunlar (`cat_cols`)**:
    
    -   Veri türü `object` olan sütunlar.
    -   Sayısal olmasına rağmen benzersiz (unique) değeri belirli bir eşik değerinin (`cat_th`) altında olan sütunlar.
2.  **Sayısal Sütunlar (`num_cols`)**:
    
    -   Veri türü sayısal (`int` veya `float`) olan sütunlar.
3.  **Kategorik Ama Kardinal Sütunlar (`cat_but_car`)**:
    
    -   Veri türü `object` olan, ancak benzersiz değer sayısı belirli bir eşik değerinin (`car_th`) üzerinde olan sütunlar.

----------

## **📊 Fonksiyonda Yapılan İşlemler**

1.  **Kategorik Sütunları Tespit Etme:**
    
    -   `object` türündeki sütunlar ve benzersiz değer sayısı az olan sayısal sütunlar (`num_but_cat`) birleştirilir.
    -   Kardinal kategorik sütunlar (`cat_but_car`) bu listeden çıkarılır.
2.  **Sayısal Sütunları Tespit Etme:**
    
    -   `int` veya `float` türündeki sütunlar belirlenir.
    -   Benzersiz değer sayısı düşük olan sütunlar (`num_but_cat`) bu listeden çıkarılır.
3.  **Kardinal Kategorik Sütunları Tespit Etme:**
    
    -   `object` türünde olup benzersiz değer sayısı yüksek olan sütunlar bu kategoriye dahil edilir.
4.  **Bilgi Çıktısı:**
    
    -   Veri çerçevesindeki toplam gözlem ve sütun sayısı.
    -   Her sütun türü için tespit edilen sütun sayısı ekrana yazdırılır.
5.  **Dönen Değerler:**
    
    -   Kategorik sütunlar (`cat_cols`).
    -   Sayısal sütunlar (`num_cols`).
    -   Kardinal kategorik sütunlar (`cat_but_car`).

-   **Sayısal Sütunlar:** Sayısal türde olan sütunlar çıkarılmıştır.
-   **Kardinal Kategorik Sütunlar:** Çok sayıda benzersiz değere sahip kategorik sütunlar analiz edilmiştir.

```python
def grab_col_names(dataframe, cat_th=10, car_th=20):
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes =="O"]
    num_but_cot = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                   dataframe[col].dtypes != "O"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and
                   dataframe[col].dtypes == "O"]
    cat_cols = cat_cols + num_but_cot
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cot]
    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f"cat_cols: {len(cat_cols)}")
    print(f"num_cols: {len(num_cols)}")
    print(f"cat_but_car: {len(cat_but_car)}")
    print(f"num_but_cat: {len(num_but_cot)}")
    return cat_cols, num_cols, cat_but_car

cat_cols, num_cols, cat_but_car = grab_col_names(crime_2020_to_present_data)
```

# 🛠️ **Eksik Verileri Belirli Sütunlarda 0 ile Doldurma**

Bu kod, yalnızca veri çerçevesinde mevcut olan belirli sütunlardaki eksik değerleri (`NaN`) **0** ile doldurmayı amaçlamaktadır. Bu işlem, özellikle modelleme süreçlerinde eksik verilerin olumsuz etkilerini azaltmak için kullanılan yaygın bir veri temizleme yöntemidir.

----------

## **🔍 Yapılan İşlemler**

3.  **Eksik Değerleri Doldurma:**

# 🔄 **Feature Engineering: Tarih Bilgilerinin Dönüştürülmesi ve Zenginleştirilmesi**

Bu kod, veri setindeki tarih sütunlarının **datetime formatına dönüştürülmesi** ve bu sütunlardan **yıl, ay, gün** gibi ayrıntılı zaman bilgileri çıkarılması işlemlerini içerir. Bu adım, makine öğrenimi ve veri analizi süreçlerinde tarih bilgilerini daha verimli kullanmak için önemli bir **feature engineering** tekniğidir.

----------

## **🔍 Yapılan İşlemler**

### 1. **Tarih Sütunlarını Dönüştürme:**

-   **`Date Rptd`** ve **`DATE OCC`** sütunları, **datetime formatına** dönüştürülmüştür.
-   Sütunlar, yalnızca **tarih** bilgisiyle (saat ve dakika bilgisi olmadan) güncellenmiştir.
-   Bu işlem için **`pd.to_datetime()`** fonksiyonu kullanılmıştır.
    -   **`format='%m/%d/%Y %I:%M:%S %p'`**: Tarih formatı ayarlanmıştır.

### 2. **Yıl, Ay ve Gün Bilgilerinin Çıkarılması:**

-   Dönüştürülen **`Date Rptd`** ve **`DATE OCC`** sütunlarından:
    -   **Yıl (`_Year`)**,
    -   **Ay (`_Month`)**,
    -   **Gün (`_Day`)** bilgileri ayrı sütunlar olarak veri setine eklenmiştir.
-   Bu işlem için **`.dt`** fonksiyonu ile **`year`, `month` ve `day`** bilgileri alınmıştır.
-   Her tarih sütunundan yeni özellikler türetilerek veri zenginleştirilmiştir.

### 3. **Güncellenmiş Veriyi İnceleme:**

-   Veri setinin ilk beş satırı **`head(5)`** ile yazdırılarak yapılan değişiklikler kontrol edilmiştir.

```python
# Tarih sütunlarını datetime formatına çeviriyoruz (saat ve dakika bilgileri olmadan)
crime_2020_to_present_data['Date Rptd'] = pd.to_datetime(crime_2020_to_present_data['Date Rptd'], format='%m/%d/%Y %I:%M:%S %p').dt.date
crime_2020_to_present_data['DATE OCC'] = pd.to_datetime(crime_2020_to_present_data['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p').dt.date

# Date Rptd ve DATE OCC sütunlarından yıl, ay, gün bilgilerini çıkarma (saat ve dakika bilgileri eklenmedi)
for col in ['Date Rptd', 'DATE OCC']:
    crime_2020_to_present_data[f'{col}_Year'] = pd.to_datetime(crime_2020_to_present_data[col]).dt.year
    crime_2020_to_present_data[f'{col}_Month'] = pd.to_datetime(crime_2020_to_present_data[col]).dt.month
    crime_2020_to_present_data[f'{col}_Day'] = pd.to_datetime(crime_2020_to_present_data[col]).dt.day

# Veri setinin güncellenmiş ilk beş satırını yazdırma
crime_2020_to_present_data.head(5)
```

# 📅 **Haftanın Günlerine Göre Los Angeles Suç Dağılımı**

Bu kod, Los Angeles suç verisindeki tarih bilgilerini kullanarak suçların haftanın günlerine göre dağılımını analiz eder ve modern bir yatay çubuk grafiği ile görselleştirir. Haftalık dağılım, suçların hangi günlerde yoğunlaştığını anlamak için önemli bir adımdır.

----------

## **🔍 Yapılan İşlemler**

### 1. **Tarih Bilgilerinden Haftanın Günlerini Elde Etme:**

-   **`DATE OCC`** sütunu, tarih formatına dönüştürülerek doğrulandı.
-   **`.dt.day_name()`** kullanılarak, her tarihin haftanın hangi gününe denk geldiği tespit edildi ve yeni bir sütun olan **`Weekday`** eklendi.

### 2. **Haftanın Günlerine Göre Suç Sayısının Hesaplanması:**

-   **`value_counts()`** fonksiyonu ile her gün için suç sayıları hesaplandı.
-   **`reindex()`** ile günler kronolojik sıraya (**Monday, Tuesday, ...**) göre düzenlendi.

### 3. **Görselleştirme: Yatay Çubuk Grafiği**

-   **`sns.barplot()`** ile suçların haftanın günlerine göre dağılımını gösteren bir yatay çubuk grafiği oluşturuldu.
    -   Modern ve dikkat çekici bir görünüm için **`viridis`** renk paleti kullanıldı.
    -   Barlar üzerine suç sayıları eklenerek veriler görselleştirildi.
    -   Binlik ayrım virgülleri ile sayı formatı daha okunabilir hale getirildi.

### 4. **Grafik Özelleştirme:**

-   Başlık, eksen etiketleri ve yazı boyutları profesyonel ve okunaklı bir şekilde ayarlandı.
-   Grafik çerçevesi temizlenerek gereksiz sınırlar kaldırıldı (**`spines['top']`** ve **`spines['right']`** gizlendi).
-   Daha modern bir görünüm için:
    -   Grafik ve arka plan renkleri beyaz yapıldı.
    -   Grid çizgileri eklenerek grafik okunabilirliği artırıldı.

```python
# DATE OCC kolonundan haftanın günlerini elde etme
crime_2020_to_present_data['DATE OCC'] = pd.to_datetime(crime_2020_to_present_data['DATE OCC'])  # Tarih formatına çevirme
crime_2020_to_present_data['Weekday'] = crime_2020_to_present_data['DATE OCC'].dt.day_name()  # Haftanın günü olarak yeni bir sütun ekleme

# Haftanın günlerine göre suç sayısını hesaplama
crime_counts = crime_2020_to_present_data['Weekday'].value_counts().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Görselleştirme
plt.figure(figsize=(16, 9))

# Kullanılan renk paletini daha modern ve uyumlu yapma
colors = sns.color_palette("viridis", n_colors=7)

# Haftanın günlerine göre suçları sayma ve soldan sağa yatay bar grafiği oluşturma
ax = sns.barplot(x=crime_counts.values, y=crime_counts.index, palette=colors)

# Başlık ve etiketler
plt.title('Haftanın Günlerine Göre Los Angeles Suç Dağılımı', fontsize=24, fontweight='bold', color='#333333', pad=20)
plt.xlabel('Olay Sayısı', fontsize=18, fontweight='bold', labelpad=15)
plt.ylabel('Hafta Günleri', fontsize=18, fontweight='bold', labelpad=15)

# X ve Y etiketlerini daha sade ve profesyonel yapma
plt.xticks(fontsize=14, fontweight='medium', color='#333333')
plt.yticks(fontsize=16, fontweight='medium', color='#333333')

# X eksenindeki sayıların binlik ayrımını virgülle gösterme
ax.xaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{int(x):,}'))

# Her bir bar üzerine değerleri ekleyerek daha sade ve etkileyici hale getirme
for p in ax.patches:
    ax.annotate(f'{int(p.get_width()):,}', 
                (p.get_width() + 500, p.get_y() + p.get_height() / 2),  # Barın sağında yer alacak şekilde konumlandırma
                ha='left', va='center', fontsize=14, color='black', xytext=(5, 0),
                textcoords='offset points')

# Grafik sınırlarını daha belirgin yapma
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color('#CCCCCC')
ax.spines['left'].set_color('#CCCCCC')

# Arka planı beyaz yaparak daha temiz ve modern bir görünüm elde etme
plt.gcf().set_facecolor('white')
ax.set_facecolor('#F9F9F9')  # Grafiğin iç arka plan rengi

# Grid çizgilerini ekleme ve saydam yapma
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show() 
```

# 🔢 **Suç Kodu Sayısını Hesaplama ve Yeni Özellik Ekleme**

Bu kod, her satırda bulunan suç kodlarının sayısını hesaplayarak veri setine yeni bir özellik (sütun) ekler. Bu tür özellik mühendisliği, her olayın karmaşıklığını veya birden fazla suçun aynı anda işlendiği durumları analiz etmek için kullanışlıdır.

----------

## **🔍 Yapılan İşlemler**

### 1. **Suç Kodu Sütunlarını Belirleme:**

-   **`crime_code_columns`** değişkeni, suç kodlarının bulunduğu sütunları tanımlar:
    -   `"Crm Cd 1"`
    -   `"Crm Cd 2"`
    -   `"Crm Cd 3"`

### 2. **Her Satırda Suç Kodu Sayısını Hesaplama:**

-   **`.apply()`** fonksiyonu ile her satırın belirtilen sütunlarında işlem yapılmıştır:
    -   **`row != 0`**: Her satırda sıfırdan farklı (dolu) olan sütunları kontrol eder.
    -   **`.sum()`**: Dolu sütunların toplam sayısını hesaplar.
-   Bu işlem, bir olayda kaç farklı suç kodunun bulunduğunu hesaplar.

### 3. **Yeni Sütun Eklenmesi:**

-   Hesaplanan değerler, veri setine **`Crime Code Count`** isimli yeni bir sütun olarak eklenmiştir.
-   Bu sütun, her olayın içerdiği suç kodu sayısını temsil eder.

-   **`head()`** kullanılarak yeni sütunun doğruluğu kontrol edilmiş ve suç kodları ile karşılaştırılmıştır.

```python
# Suç kodu sütunlarını belirtiyoruz
crime_code_columns = ["Crm Cd 1", "Crm Cd 2", "Crm Cd 3"]

# Yeni sütun ekliyoruz ve her satırda dolu olan suç kodlarını sayıyoruz
crime_2020_to_present_data['Crime Code Count'] = crime_2020_to_present_data[crime_code_columns].apply(lambda row: (row != 0).sum(), axis=1) 
# Sonucu kontrol edelim
crime_2020_to_present_data[["Crm Cd 1", "Crm Cd 2", "Crm Cd 3", "Crime Code Count"]]

```

# 🏷️ **One-Hot Encoding ile Kategorik Verilerin Dönüştürülmesi**

Bu kod, veri setindeki belirli kategorik sütunları **One-Hot Encoding** yöntemiyle sayısal değerlere dönüştürür. Bu işlem, makine öğrenimi modellerinin kategorik verilerle daha iyi çalışmasını sağlar.

----------

## **🔍 Yapılan İşlemler**

### 1. **Dönüştürülecek Sütunları Tanımlama:**

-   **`one_hot_columns`** değişkeni ile One-Hot Encoding uygulanacak sütunlar tanımlanmıştır:
    -   **`AREA NAME`**: Suçun gerçekleştiği bölge.
    -   **`Vict Sex`**: Mağdurun cinsiyeti.
    -   **`Vict Descent`**: Mağdurun etnik kökeni.
    -   **`Status Desc`**: Suç olayının durumu (örneğin, açık veya kapalı dava).

### 2. **One-Hot Encoding Uygulama:**

-   **`pd.get_dummies()`** fonksiyonu kullanılarak, seçilen sütunlar için One-Hot Encoding uygulanmıştır:
    -   Her benzersiz kategorik değer, veri setine yeni bir sütun olarak eklenmiştir.
    -   Sütun isimleri, kategorik değerlerin daha kolay anlaşılmasını sağlamak için orijinal sütun adlarıyla **`prefix`** parametresi kullanılarak oluşturulmuştur.

### 3. **Yeni Veri Çerçevesi:**

-   **`crime_2020_to_present_data_encoded`** isimli yeni bir veri çerçevesi oluşturulmuş, orijinal kategorik sütunlar sayısal sütunlarla değiştirilmiştir.

### 4. **Sonuçları Kontrol Etme:**

-   **`head(5)`** ile veri setinin ilk 5 satırı görüntülenmiş, dönüşümün başarılı olduğu doğrulanmıştır.

```python
# One-Hot Encoding uygulanacak uygun öznitelikleri tanımlıyoruz
one_hot_columns = ['AREA NAME', 'Vict Sex', 'Vict Descent', 'Status Desc']

# One-Hot Encoding işlemi
crime_2020_to_present_data_encoded = pd.get_dummies(crime_2020_to_present_data, columns=one_hot_columns, prefix=one_hot_columns)

# İlk 5 satırı yazdırarak sonucu kontrol ediyoruz
crime_2020_to_present_data_encoded.head(5)

```
