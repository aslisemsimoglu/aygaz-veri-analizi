# üõ†Ô∏è **Rastgele Eksik Veri Ekleme Fonksiyonu**

Bu b√∂l√ºmde, bir veri √ßer√ßevesine rastgele eksik deƒüerler eklemek i√ßin bir fonksiyon olu≈üturulmu≈ü ve uygulanmƒ±≈ütƒ±r. Eksik veri ekleme, √∂zellikle makine √∂ƒürenimi ve veri analitiƒüi projelerinde, eksik verilerle ba≈üa √ßƒ±kma y√∂ntemlerini test etmek i√ßin kullanƒ±≈ülƒ±dƒ±r.

----------

## **üîç Fonksiyonun Amacƒ±**

### **Fonksiyon: `add_random_missing_values_to_all_columns`**

Bu fonksiyonun g√∂revi, belirtilen oran (%5) doƒürultusunda bir veri √ßer√ßevesindeki t√ºm s√ºtunlara rastgele eksik (NaN) deƒüerler eklemektir.

### **Fonksiyonun Parametreleri:**

-   **`dataframe`**: √úzerinde i≈ülem yapƒ±lacak Pandas veri √ßer√ßevesi.
-   **`missing_rate_per_col`**: Her s√ºtunda ne kadar eksik veri eklenmesi gerektiƒüini belirten oran (varsayƒ±lan olarak %5).
-   **`seed`**: Rastgelelik i√ßin sabit bir ba≈ülangƒ±√ß noktasƒ± belirlemek ve tekrar edilebilir sonu√ßlar almak i√ßin kullanƒ±lƒ±r.

----------

## **üí° Yapƒ±lan ƒ∞≈ülemler**

1.  **Orijinal Veriyi Koruma:**
    
    -   Veri √ßer√ßevesinin bir kopyasƒ± olu≈üturularak orijinal veri korunur.
2.  **Rastgelelik Kontrol√º:**
    
    -   `random.seed()` ve `np.random.seed()` kullanƒ±larak rastgele i≈ülemler sabitlenir. Bu, fonksiyonun tekrar √ßalƒ±≈ütƒ±rƒ±ldƒ±ƒüƒ±nda aynƒ± sonu√ßlarƒ± √ºretmesini saƒülar.
3.  **S√ºtun T√ºrlerini D√∂n√º≈üt√ºrme:**
    
    -   Eksik deƒüerler (NaN) yalnƒ±zca float t√ºr√ºndeki verilere atanabileceƒüi i√ßin, integer t√ºr√ºndeki s√ºtunlar float‚Äôa d√∂n√º≈üt√ºr√ºl√ºr.
4.  **Eksik Veri Ekleme:**
    
    -   Her s√ºtun i√ßin:
        -   Eksik veri eklenmesi gereken satƒ±r sayƒ±sƒ± hesaplanƒ±r.
        -   Rastgele se√ßilen satƒ±rlarƒ±n indekslerine NaN atanƒ±r.

----------

## **üìä Uygulama: Los Angeles Su√ß Verisi**

Fonksiyon, **Los Angeles su√ß verisi** √ºzerinde ≈üu ≈üekilde uygulanmƒ±≈ütƒ±r:

-   Her s√ºtuna toplam verinin %5'i kadar eksik veri eklenmi≈ütir.
-   Bu adƒ±m, eksik veri analizi ve doldurma y√∂ntemlerini test etmek i√ßin kullanƒ±≈ülƒ±dƒ±r.

```python
def add_random_missing_values_to_all_columns(dataframe: pd.DataFrame,
                                             missing_rate_per_col: float = 0.05,
                                             seed: int = 42) -> pd.DataFrame:
    """Adds random missing values to every column in a DataFrame.
    
    Args:
        dataframe (pd.DataFrame): DataFrame to be processed.
        missing_rate_per_col (float): Percentage of missing value rate in float format for each column. Defaults 0.05
        seed (int): Random seed for reproducibility.
        
    Returns:
        pd.DataFrame: DataFrame with added missing values.
    """
    # Copying the dataframe to prevent changing the original data
    df_missing = dataframe.copy()

    # Set seed for reproducibility
    random.seed(seed)
    np.random.seed(seed)

    # Convert integer columns to float to allow NaN values
    for col in df_missing.select_dtypes(include=['int']).columns:
        df_missing[col] = df_missing[col].astype(float)

    # Iterate over each column to add missing values
    for col in df_missing.columns:
        # Number of missing values to add in this column
        num_missing = int(len(df_missing) * missing_rate_per_col)
        
        # Generate random row indices to set as NaN
        missing_indices = np.random.choice(df_missing.index, num_missing, replace=False)
        
        # Set those indices to NaN in the respective column
        df_missing.loc[missing_indices, col] = np.nan

    return df_missing

# Veri setine her s√ºtuna eksik veri eklemek i√ßin
crime_2020_to_present_data = add_random_missing_values_to_all_columns(
    dataframe=crime_2020_to_present_data,
    missing_rate_per_col=0.05  # Her s√ºtuna %5 eksik veri ekleniyor
)
```

# **üîç Veri √áer√ßevesi Kontrol Fonksiyonu: `check_df`**

Bu b√∂l√ºmde, bir veri √ßer√ßevesini detaylƒ± ≈üekilde analiz etmek ve d√ºzenli bir bi√ßimde g√∂r√ºnt√ºlemek i√ßin bir fonksiyon olu≈üturulmu≈ütur. **`check_df`**, veri √ßer√ßevesi hakkƒ±nda temel bilgileri, eksik deƒüerleri ve istatistiksel √∂zetleri kullanƒ±cƒ± dostu bir ≈üekilde sunar. üìä‚ú®

----------

## **üõ†Ô∏è Fonksiyonun Amacƒ± ve Yapƒ±lan ƒ∞≈ülemler**

### üéØ **Fonksiyonun Amacƒ±**

`check_df` fonksiyonu, bir veri √ßer√ßevesinin:

-   Boyutlarƒ±nƒ±,
-   S√ºtun t√ºrlerini,
-   ƒ∞lk ve son birka√ß satƒ±rƒ±nƒ±,
-   Eksik deƒüerlerini,
-   ƒ∞statistiksel √∂zetini g√∂r√ºnt√ºlemeyi saƒülar.

### üîß **Fonksiyonda Yapƒ±lan ƒ∞≈ülemler:**

1.  **Genel Bilgiler (`info`)**
    
    -   Veri √ßer√ßevesi hakkƒ±nda s√ºtun bazƒ±nda **non-null (bo≈ü olmayan)** deƒüer sayƒ±sƒ± ve veri t√ºrleri (`dtype`) gibi bilgiler tablola≈ütƒ±rƒ±larak d√ºzenli bir ≈üekilde g√∂sterildi.
2.  **Boyut Bilgisi (`shape`)**
    
    -   Veri √ßer√ßevesinin satƒ±r ve s√ºtun sayƒ±sƒ± kullanƒ±cƒ±ya bildirildi.
3.  **S√ºtun T√ºrleri (`dtypes`)**
    
    -   T√ºm s√ºtunlarƒ±n veri t√ºrleri (`int`, `float`, `object` gibi) g√∂r√ºnt√ºlendi.
4.  **ƒ∞lk ve Son Satƒ±rlar (`head` ve `tail`)**
    
    -   Veri √ßer√ßevesinin ilk ve son birka√ß satƒ±rƒ± g√∂sterilerek genel yapƒ± hakkƒ±nda bilgi saƒülandƒ±.
5.  **Eksik Deƒüerler (`isnull`)**
    
    -   S√ºtun bazƒ±nda eksik deƒüerlerin toplamƒ± hesaplanƒ±p d√ºzenli bir ≈üekilde g√∂r√ºnt√ºlendi.
6.  **ƒ∞statistiksel √ñzet (`describe`)**
    
    -   Sayƒ±sal s√ºtunlara ait istatistiksel √∂zetler (ortalama, standart sapma, min, max gibi) tablola≈ütƒ±rƒ±ldƒ±.

----------

## **üìä Fonksiyonun Uygulamasƒ±: Los Angeles Su√ß Verisi**

-   **Veri Y√ºkleme:** `pd.read_csv()` ile **Los Angeles su√ß verisi** y√ºklendi.
-   **Fonksiyon Kullanƒ±mƒ±:** `check_df` fonksiyonu kullanƒ±larak veri √ßer√ßevesine dair:
    -   Yapƒ±sal bilgiler,
    -   Eksik deƒüer analizi,
    -   Sayƒ±sal √∂zetler kolayca g√∂r√ºnt√ºlendi.

```python
def check_df(dataframe, head=5):
    # Info - D√ºzenli tablo halinde g√∂sterim
    buffer = StringIO()
    dataframe.info(buf=buffer)
    info_str = buffer.getvalue()

    # Info'yu tablola≈ütƒ±rma
    info_lines = info_str.splitlines()
    col_info = []
    columns = ['#', 'Column', 'Non-Null Count', 'Dtype']  # S√ºtun isimlerini dinamik hale getiriyoruz

    for line in info_lines[5:]:
        if line.strip():
            cols = line.split(maxsplit=3)  # Sadece 4 s√ºtun ayƒ±rarak fazla b√∂l√ºnmeyi engelliyoruz
            if len(cols) == 4:  # Eƒüer t√ºm s√ºtunlar varsa ekliyoruz
                col_info.append(cols)

    # DataFrame olu≈üturma ve g√∂r√ºnt√ºleme
    info_df = pd.DataFrame(col_info, columns=columns)
    display(HTML("<h3>Info</h3>"))
    display(info_df)

    # Shape
    display(HTML(f"<h3>Shape</h3><p>{dataframe.shape}</p>"))
    
    # Types
    display(HTML("<h3>Types</h3>"))
    types_df = pd.DataFrame({'Column': dataframe.columns, 'Dtype': dataframe.dtypes}).transpose()
    display(types_df)
    
    # Head
    display(HTML("<h3>Head</h3>"))
    display(dataframe.head(head))
    
    # Tail
    display(HTML("<h3>Tail</h3>"))
    display(dataframe.tail(head))
    
    # NA (Null Values)
    display(HTML("<h3>NA (Null Values)</h3>"))
    na_df = dataframe.isnull().sum().to_frame().T
    na_df.columns.name = None  # S√ºtun isimlerinden `name` √∂zelliƒüini temizledik
    display(na_df)
    
    # Describe
    display(HTML("<h3>Describe</h3>"))
    describe_df = dataframe.describe().transpose()
    describe_df.columns.name = None  # S√ºtun isimlerinden `name` √∂zelliƒüini temizledik
    display(describe_df)

# Crime data √ßer√ßevesini kontrol etme
crime_2020_to_present_data = pd.read_csv("/kaggle/input/los-angeles-crime-data-from-2000/Crime_Data_from_2020_to_Present.csv")
check_df(crime_2020_to_present_data)
```

# üõ†Ô∏è **S√ºtun T√ºrlerini Otomatik Tespit Etme: `grab_col_names`**

Bu fonksiyon, bir veri √ßer√ßevesindeki s√ºtunlarƒ± **kategorik**, **sayƒ±sal**, ve √∂zel kategorilere ayƒ±rmak i√ßin geli≈ütirilmi≈ütir. Veri analizi ve modelleme s√ºre√ßlerinde, s√ºtun t√ºrlerini belirlemek √∂nemli bir adƒ±mdƒ±r ve bu fonksiyon bu i≈ülemi otomatikle≈ütirir. üîç‚ú®

----------

## **üîç Fonksiyonun Amacƒ±**

**`grab_col_names`**, veri √ßer√ßevesindeki s√ºtunlarƒ± √∂zelliklerine g√∂re √º√ß ana kategoriye ayƒ±rƒ±r:

1.  **Kategorik S√ºtunlar (`cat_cols`)**:
    
    -   Veri t√ºr√º `object` olan s√ºtunlar.
    -   Sayƒ±sal olmasƒ±na raƒümen benzersiz (unique) deƒüeri belirli bir e≈üik deƒüerinin (`cat_th`) altƒ±nda olan s√ºtunlar.
2.  **Sayƒ±sal S√ºtunlar (`num_cols`)**:
    
    -   Veri t√ºr√º sayƒ±sal (`int` veya `float`) olan s√ºtunlar.
3.  **Kategorik Ama Kardinal S√ºtunlar (`cat_but_car`)**:
    
    -   Veri t√ºr√º `object` olan, ancak benzersiz deƒüer sayƒ±sƒ± belirli bir e≈üik deƒüerinin (`car_th`) √ºzerinde olan s√ºtunlar.

----------

## **üìä Fonksiyonda Yapƒ±lan ƒ∞≈ülemler**

1.  **Kategorik S√ºtunlarƒ± Tespit Etme:**
    
    -   `object` t√ºr√ºndeki s√ºtunlar ve benzersiz deƒüer sayƒ±sƒ± az olan sayƒ±sal s√ºtunlar (`num_but_cat`) birle≈ütirilir.
    -   Kardinal kategorik s√ºtunlar (`cat_but_car`) bu listeden √ßƒ±karƒ±lƒ±r.
2.  **Sayƒ±sal S√ºtunlarƒ± Tespit Etme:**
    
    -   `int` veya `float` t√ºr√ºndeki s√ºtunlar belirlenir.
    -   Benzersiz deƒüer sayƒ±sƒ± d√º≈ü√ºk olan s√ºtunlar (`num_but_cat`) bu listeden √ßƒ±karƒ±lƒ±r.
3.  **Kardinal Kategorik S√ºtunlarƒ± Tespit Etme:**
    
    -   `object` t√ºr√ºnde olup benzersiz deƒüer sayƒ±sƒ± y√ºksek olan s√ºtunlar bu kategoriye dahil edilir.
4.  **Bilgi √áƒ±ktƒ±sƒ±:**
    
    -   Veri √ßer√ßevesindeki toplam g√∂zlem ve s√ºtun sayƒ±sƒ±.
    -   Her s√ºtun t√ºr√º i√ßin tespit edilen s√ºtun sayƒ±sƒ± ekrana yazdƒ±rƒ±lƒ±r.
5.  **D√∂nen Deƒüerler:**
    
    -   Kategorik s√ºtunlar (`cat_cols`).
    -   Sayƒ±sal s√ºtunlar (`num_cols`).
    -   Kardinal kategorik s√ºtunlar (`cat_but_car`).

----------

## **üìã Uygulama: Los Angeles Su√ß Verisi**

Fonksiyon, **Los Angeles su√ß verisi** √ºzerinde uygulanmƒ±≈ü ve s√ºtun t√ºrleri a≈üaƒüƒ±daki gibi ayrƒ±lmƒ±≈ütƒ±r:

-   **Kategorik S√ºtunlar:** Hangi deƒüi≈ükenlerin kategorik olduƒüu belirlenmi≈ütir.
-   **Sayƒ±sal S√ºtunlar:** Sayƒ±sal t√ºrde olan s√ºtunlar √ßƒ±karƒ±lmƒ±≈ütƒ±r.
-   **Kardinal Kategorik S√ºtunlar:** √áok sayƒ±da benzersiz deƒüere sahip kategorik s√ºtunlar analiz edilmi≈ütir.

```python
def grab_col_names(dataframe, cat_th=10, car_th=20):
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes =="O"]
    num_but_cot = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                   dataframe[col].dtypes != "O"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and
                   dataframe[col].dtypes == "O"]
    cat_cols = cat_cols + num_but_cot
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cot]
    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f"cat_cols: {len(cat_cols)}")
    print(f"num_cols: {len(num_cols)}")
    print(f"cat_but_car: {len(cat_but_car)}")
    print(f"num_but_cat: {len(num_but_cot)}")
    return cat_cols, num_cols, cat_but_car

cat_cols, num_cols, cat_but_car = grab_col_names(crime_2020_to_present_data)
```

# üîÑ **Feature Engineering: Tarih Bilgilerinin D√∂n√º≈üt√ºr√ºlmesi ve Zenginle≈ütirilmesi**

Bu kod, veri setindeki tarih s√ºtunlarƒ±nƒ±n **datetime formatƒ±na d√∂n√º≈üt√ºr√ºlmesi** ve bu s√ºtunlardan **yƒ±l, ay, g√ºn** gibi ayrƒ±ntƒ±lƒ± zaman bilgileri √ßƒ±karƒ±lmasƒ± i≈ülemlerini i√ßerir. Bu adƒ±m, makine √∂ƒürenimi ve veri analizi s√ºre√ßlerinde tarih bilgilerini daha verimli kullanmak i√ßin √∂nemli bir **feature engineering** tekniƒüidir.

----------

## **üîç Yapƒ±lan ƒ∞≈ülemler**

### 1. **Tarih S√ºtunlarƒ±nƒ± D√∂n√º≈üt√ºrme:**

-   **`Date Rptd`** ve **`DATE OCC`** s√ºtunlarƒ±, **datetime formatƒ±na** d√∂n√º≈üt√ºr√ºlm√º≈üt√ºr.
-   S√ºtunlar, yalnƒ±zca **tarih** bilgisiyle (saat ve dakika bilgisi olmadan) g√ºncellenmi≈ütir.
-   Bu i≈ülem i√ßin **`pd.to_datetime()`** fonksiyonu kullanƒ±lmƒ±≈ütƒ±r.
    -   **`format='%m/%d/%Y %I:%M:%S %p'`**: Tarih formatƒ± ayarlanmƒ±≈ütƒ±r.

### 2. **Yƒ±l, Ay ve G√ºn Bilgilerinin √áƒ±karƒ±lmasƒ±:**

-   D√∂n√º≈üt√ºr√ºlen **`Date Rptd`** ve **`DATE OCC`** s√ºtunlarƒ±ndan:
    -   **Yƒ±l (`_Year`)**,
    -   **Ay (`_Month`)**,
    -   **G√ºn (`_Day`)** bilgileri ayrƒ± s√ºtunlar olarak veri setine eklenmi≈ütir.
-   Bu i≈ülem i√ßin **`.dt`** fonksiyonu ile **`year`, `month` ve `day`** bilgileri alƒ±nmƒ±≈ütƒ±r.
-   Her tarih s√ºtunundan yeni √∂zellikler t√ºretilerek veri zenginle≈ütirilmi≈ütir.

### 3. **G√ºncellenmi≈ü Veriyi ƒ∞nceleme:**

-   Veri setinin ilk be≈ü satƒ±rƒ± **`head(5)`** ile yazdƒ±rƒ±larak yapƒ±lan deƒüi≈üiklikler kontrol edilmi≈ütir.

```python
# Tarih s√ºtunlarƒ±nƒ± datetime formatƒ±na √ßeviriyoruz (saat ve dakika bilgileri olmadan)
crime_2020_to_present_data['Date Rptd'] = pd.to_datetime(crime_2020_to_present_data['Date Rptd'], format='%m/%d/%Y %I:%M:%S %p').dt.date
crime_2020_to_present_data['DATE OCC'] = pd.to_datetime(crime_2020_to_present_data['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p').dt.date

# Date Rptd ve DATE OCC s√ºtunlarƒ±ndan yƒ±l, ay, g√ºn bilgilerini √ßƒ±karma (saat ve dakika bilgileri eklenmedi)
for col in ['Date Rptd', 'DATE OCC']:
    crime_2020_to_present_data[f'{col}_Year'] = pd.to_datetime(crime_2020_to_present_data[col]).dt.year
    crime_2020_to_present_data[f'{col}_Month'] = pd.to_datetime(crime_2020_to_present_data[col]).dt.month
    crime_2020_to_present_data[f'{col}_Day'] = pd.to_datetime(crime_2020_to_present_data[col]).dt.day

# Veri setinin g√ºncellenmi≈ü ilk be≈ü satƒ±rƒ±nƒ± yazdƒ±rma
crime_2020_to_present_data.head(5)
```

```python
# DATE OCC kolonundan haftanƒ±n g√ºnlerini elde etme
crime_2020_to_present_data['DATE OCC'] = pd.to_datetime(crime_2020_to_present_data['DATE OCC'])  # Tarih formatƒ±na √ßevirme
crime_2020_to_present_data['Weekday'] = crime_2020_to_present_data['DATE OCC'].dt.day_name()  # Haftanƒ±n g√ºn√º olarak yeni bir s√ºtun ekleme

# Haftanƒ±n g√ºnlerine g√∂re su√ß sayƒ±sƒ±nƒ± hesaplama
crime_counts = crime_2020_to_present_data['Weekday'].value_counts().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# G√∂rselle≈ütirme
plt.figure(figsize=(16, 9))

# Kullanƒ±lan renk paletini daha modern ve uyumlu yapma
colors = sns.color_palette("viridis", n_colors=7)

# Haftanƒ±n g√ºnlerine g√∂re su√ßlarƒ± sayma ve soldan saƒüa yatay bar grafiƒüi olu≈üturma
ax = sns.barplot(x=crime_counts.values, y=crime_counts.index, palette=colors)

# Ba≈ülƒ±k ve etiketler
plt.title('Haftanƒ±n G√ºnlerine G√∂re Los Angeles Su√ß Daƒüƒ±lƒ±mƒ±', fontsize=24, fontweight='bold', color='#333333', pad=20)
plt.xlabel('Olay Sayƒ±sƒ±', fontsize=18, fontweight='bold', labelpad=15)
plt.ylabel('Hafta G√ºnleri', fontsize=18, fontweight='bold', labelpad=15)

# X ve Y etiketlerini daha sade ve profesyonel yapma
plt.xticks(fontsize=14, fontweight='medium', color='#333333')
plt.yticks(fontsize=16, fontweight='medium', color='#333333')

# X eksenindeki sayƒ±larƒ±n binlik ayrƒ±mƒ±nƒ± virg√ºlle g√∂sterme
ax.xaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{int(x):,}'))

# Her bir bar √ºzerine deƒüerleri ekleyerek daha sade ve etkileyici hale getirme
for p in ax.patches:
    ax.annotate(f'{int(p.get_width()):,}', 
                (p.get_width() + 500, p.get_y() + p.get_height() / 2),  # Barƒ±n saƒüƒ±nda yer alacak ≈üekilde konumlandƒ±rma
                ha='left', va='center', fontsize=14, color='black', xytext=(5, 0),
                textcoords='offset points')

# Grafik sƒ±nƒ±rlarƒ±nƒ± daha belirgin yapma
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color('#CCCCCC')
ax.spines['left'].set_color('#CCCCCC')

# Arka planƒ± beyaz yaparak daha temiz ve modern bir g√∂r√ºn√ºm elde etme
plt.gcf().set_facecolor('white')
ax.set_facecolor('#F9F9F9')  # Grafiƒüin i√ß arka plan rengi

# Grid √ßizgilerini ekleme ve saydam yapma
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show() 
```

# üî¢ **Su√ß Kodu Sayƒ±sƒ±nƒ± Hesaplama ve Yeni √ñzellik Ekleme**

Bu kod, her satƒ±rda bulunan su√ß kodlarƒ±nƒ±n sayƒ±sƒ±nƒ± hesaplayarak veri setine yeni bir √∂zellik (s√ºtun) ekler. Bu t√ºr √∂zellik m√ºhendisliƒüi, her olayƒ±n karma≈üƒ±klƒ±ƒüƒ±nƒ± veya birden fazla su√ßun aynƒ± anda i≈ülendiƒüi durumlarƒ± analiz etmek i√ßin kullanƒ±≈ülƒ±dƒ±r.

----------

## **üîç Yapƒ±lan ƒ∞≈ülemler**

### 1. **Su√ß Kodu S√ºtunlarƒ±nƒ± Belirleme:**

-   **`crime_code_columns`** deƒüi≈ükeni, su√ß kodlarƒ±nƒ±n bulunduƒüu s√ºtunlarƒ± tanƒ±mlar:
    -   `"Crm Cd 1"`
    -   `"Crm Cd 2"`
    -   `"Crm Cd 3"`

### 2. **Her Satƒ±rda Su√ß Kodu Sayƒ±sƒ±nƒ± Hesaplama:**

-   **`.apply()`** fonksiyonu ile her satƒ±rƒ±n belirtilen s√ºtunlarƒ±nda i≈ülem yapƒ±lmƒ±≈ütƒ±r:
    -   **`row != 0`**: Her satƒ±rda sƒ±fƒ±rdan farklƒ± (dolu) olan s√ºtunlarƒ± kontrol eder.
    -   **`.sum()`**: Dolu s√ºtunlarƒ±n toplam sayƒ±sƒ±nƒ± hesaplar.
-   Bu i≈ülem, bir olayda ka√ß farklƒ± su√ß kodunun bulunduƒüunu hesaplar.

### 3. **Yeni S√ºtun Eklenmesi:**

-   Hesaplanan deƒüerler, veri setine **`Crime Code Count`** isimli yeni bir s√ºtun olarak eklenmi≈ütir.
-   Bu s√ºtun, her olayƒ±n i√ßerdiƒüi su√ß kodu sayƒ±sƒ±nƒ± temsil eder.

### 4. **Sonu√ßlarƒ±n Kontrol√º:**

-   **`head()`** kullanƒ±larak yeni s√ºtunun doƒüruluƒüu kontrol edilmi≈ü ve su√ß kodlarƒ± ile kar≈üƒ±la≈ütƒ±rƒ±lmƒ±≈ütƒ±r.

```python
# Su√ß kodu s√ºtunlarƒ±nƒ± belirtiyoruz
crime_code_columns = ["Crm Cd 1", "Crm Cd 2", "Crm Cd 3"]

# Yeni s√ºtun ekliyoruz ve her satƒ±rda dolu olan su√ß kodlarƒ±nƒ± sayƒ±yoruz
crime_2020_to_present_data['Crime Code Count'] = crime_2020_to_present_data[crime_code_columns].apply(lambda row: (row != 0).sum(), axis=1) 
# Sonucu kontrol edelim
crime_2020_to_present_data[["Crm Cd 1", "Crm Cd 2", "Crm Cd 3", "Crime Code Count"]]

```

# üè∑Ô∏è **One-Hot Encoding ile Kategorik Verilerin D√∂n√º≈üt√ºr√ºlmesi**

Bu kod, veri setindeki belirli kategorik s√ºtunlarƒ± **One-Hot Encoding** y√∂ntemiyle sayƒ±sal deƒüerlere d√∂n√º≈üt√ºr√ºr. Bu i≈ülem, makine √∂ƒürenimi modellerinin kategorik verilerle daha iyi √ßalƒ±≈ümasƒ±nƒ± saƒülar.

----------

## **üîç Yapƒ±lan ƒ∞≈ülemler**

### 1. **D√∂n√º≈üt√ºr√ºlecek S√ºtunlarƒ± Tanƒ±mlama:**

-   **`one_hot_columns`** deƒüi≈ükeni ile One-Hot Encoding uygulanacak s√ºtunlar tanƒ±mlanmƒ±≈ütƒ±r:
    -   **`AREA NAME`**: Su√ßun ger√ßekle≈ütiƒüi b√∂lge.
    -   **`Vict Sex`**: Maƒüdurun cinsiyeti.
    -   **`Vict Descent`**: Maƒüdurun etnik k√∂keni.
    -   **`Status Desc`**: Su√ß olayƒ±nƒ±n durumu (√∂rneƒüin, a√ßƒ±k veya kapalƒ± dava).

### 2. **One-Hot Encoding Uygulama:**

-   **`pd.get_dummies()`** fonksiyonu kullanƒ±larak, se√ßilen s√ºtunlar i√ßin One-Hot Encoding uygulanmƒ±≈ütƒ±r:
    -   Her benzersiz kategorik deƒüer, veri setine yeni bir s√ºtun olarak eklenmi≈ütir.
    -   S√ºtun isimleri, kategorik deƒüerlerin daha kolay anla≈üƒ±lmasƒ±nƒ± saƒülamak i√ßin orijinal s√ºtun adlarƒ±yla **`prefix`** parametresi kullanƒ±larak olu≈üturulmu≈ütur.

### 3. **Yeni Veri √áer√ßevesi:**

-   **`crime_2020_to_present_data_encoded`** isimli yeni bir veri √ßer√ßevesi olu≈üturulmu≈ü, orijinal kategorik s√ºtunlar sayƒ±sal s√ºtunlarla deƒüi≈ütirilmi≈ütir.

### 4. **Sonu√ßlarƒ± Kontrol Etme:**

-   **`head(5)`** ile veri setinin ilk 5 satƒ±rƒ± g√∂r√ºnt√ºlenmi≈ü, d√∂n√º≈ü√ºm√ºn ba≈üarƒ±lƒ± olduƒüu doƒürulanmƒ±≈ütƒ±r.

```python
# One-Hot Encoding uygulanacak uygun √∂znitelikleri tanƒ±mlƒ±yoruz
one_hot_columns = ['AREA NAME', 'Vict Sex', 'Vict Descent', 'Status Desc']

# One-Hot Encoding i≈ülemi
crime_2020_to_present_data_encoded = pd.get_dummies(crime_2020_to_present_data, columns=one_hot_columns, prefix=one_hot_columns)

# ƒ∞lk 5 satƒ±rƒ± yazdƒ±rarak sonucu kontrol ediyoruz
crime_2020_to_present_data_encoded.head(5)

```
